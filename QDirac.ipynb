```python
class Dirac():
    def __init__(self, ketString, endian='little'):
        strsize = len(ketString)
        condition1 = len(ketString)>=3
        condition2 = ketString.startswith("|")
        condition3 = ketString.endswith(">")
        condition4 = all([qubit in ["0","1"] for qubit in list(ketString[1: strsize-1])])
        condition5 = endian in ['big', 'little']
        if condition1 and condition2 and condition3 and condition4:
            pass
        else:
            raise Exception("conditons not met")

        self.QUBIT_STATE = [ele for ele in list(ketString) if not ele in ["|",">"]]
        if endian=='little':
            self.QUBIT_STATE.reverse()


        #Qubit_State = [ ele for ele in list(ketstring) if not ele in ["|",">"]]
        qb_dict = {"0": [[1.], [0.]], "1": [[0.], [1.]]}
        self.Qubit_dict = {n: qubit for n, qubit in  enumerate(self.QUBIT_STATE)}
        self.N = len(self.Qubit_dict)

        self.STATE_VECTOR = 1

        for qk, qv in self.Qubit_dict.items():
            vec = qb_dict[qv]
            self.STATE_VECTOR = matrix(kron(vec, self.STATE_VECTOR))
            
    #private method
    def _Kron(self, on_qubit,  Gate):
        #Single Qubit Identity Operation
        i2 = matrix(eye(2))
        Unitary_Operation_Matrix = {i : i2 for i in range(0, self.N)}
        if on_qubit in Unitary_Operation_Matrix:
            Unitary_Operation_Matrix[on_qubit] = Gate
        #out = reduce(kron, op_matrix)
        #self.STATE_VECTOR = matmul(out, self.STATE_VECTOR)
        return Unitary_Operation_Matrix

    #Unitary gate operations(X, Y and Z from Pauli's operations/matrices)
    def X(self, operate_on):
        i2 = matrix([[1., 0.],[0., 1.]])
        x2 = matrix([[0., 1.],[1., 0.]])
        if isinstance(operate_on, int):
            if 0 >= operate_on < self.N:
                raise ValueError("<<observe_on>> must be >=0 and < the Total number of Qubits")

            operators_list = list(self._Kron(operate_on, x2).values())
            U_op = reduce(kron, operators_list)
            self.STATE_VECTOR = matmul(U_op, self.STATE_VECTOR)

        if isinstance(operate_on, list):
            if 0 >= operate_on > self.N:
                raise ValueError("<<operate_on>> must have value between 0 and number of qubits")

    def Y(self, operate_on):
        i2 = matrix([[1., 0.],[0., 1.]])
        y2 = matrix([[0., -i],[i, 0.]])
        if isinstance(operate_on, int):
            #if 0 >= operate_on < self.N:
            #    raise ValueError("<<observe_on>> must be >=0 and < the Total number of Qubits")

            operators_list = list(self._Kron(operate_on, y2).values())
            U_op = reduce(kron, operators_list)
            self.STATE_VECTOR = matmul(U_op, self.STATE_VECTOR)

        if isinstance(operate_on, list):
            if 0 >= operate_on > self.N:
                raise ValueError("<<operate_on>> must have value between 0 and number of qubits")

    def Z(self, operate_on):
        i2 = matrix([[1., 0.],[0., 1.]])
        z2 = matrix([[1.,  0.],[0., -1.]])
        if isinstance(operate_on, int):
            #if 0 >= operate_on < self.N:
            #    raise ValueError("<<observe_on>> must be >=0 and < the Total number of Qubits")

            operators_list = list(self._Kron(operate_on, z2).values())
            U_op = reduce(kron, operators_list)
            self.STATE_VECTOR = matmul(U_op, self.STATE_VECTOR)

        if isinstance(operate_on, list):
            if 0 >= operate_on > self.N:
                raise ValueError("<<operate_on>> must have value between 0 and number of qubits")

    #Single Qubit Hadamard Operation/Unitary Matrix
    def H(self, operate_on):
        i2 = matrix([[1., 0.],[0., 1.]])
        h2 = pow(2, 0.5)* matrix([[1.,  1.],[1., -1.]])
        if isinstance(operate_on, int):
            #if 0 >= operate_on  < self.N:
            #    raise ValueError("<<observe_on>> must be >=0 and < the Total number of Qubits")

            operators_list = list(self._Kron(operate_on, h2).values())
            U_op = reduce(kron, operators_list)
            self.STATE_VECTOR = matmul(U_op, self.STATE_VECTOR)

        if isinstance(operate_on, list):
            if 0< len(operate_on) < len(self.STATE_VECTOR):
                raise ValueError("<<operate_on>> must have value between 0 and number of qubits")


    def CX(self, source:int, target:int):
        pass
```
