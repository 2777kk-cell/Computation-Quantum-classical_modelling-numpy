```python
%%writefile code/demo.c
#include <stdio.h>
#include <unistd.h>

int main(int argc, char* argv[]){

    int cherry_pick = 12;

    for(int i=0; i<16;i++){
        /*Sleep for 5 seconds*/
        if(i == cherry_pick){
            usleep(2*1000);  // Deliberately introducing delay of 25
            printf("(%d):\t %04b\n", i, i);
        }
    }
    printf("\n\n");
    return 0;
}
```

```python
import random as rand
import time
```

```python
# a list of numbers from 0 to 15 - i.e., 2**4 = 16
a_list = [i for i in range(2**4)]
a_list
```

```python
# Randomize (Shuffling list) to understand 
#worstcase scenario (while best case is obviously O(1)

rand.shuffle(a_list)

cherry_pic=12
t=0
for i in a_list:
    time.sleep(2) # Sleep of 2 seconds deliberately added to demo purpose only
    if i == cherry_pic:
        print(f"i value is {i}")
    else:
        print(f"{t} seconds lapsed")
    t+=2
        
```
